```python
import pandas as pd
```

```python
data = pd.read_csv('data/corona_tested_individuals_ver_0083_september_november.english.csv')
data.head()
```

```python
data = data.drop(['test_date'], axis = 1)
data
```

```python
df = data.copy()
for i in df.columns:
    df = df[df[i] != 'None']
df[['female', 'male']] = pd.get_dummies(data['gender'])[['female', 'male']]
df.drop('gender', axis = 1, inplace = True)
df
```

```python
df['age_60_and_above']
```

```python
df['test_indication'] = df['test_indication'].map({'Contact with confirmed': 1, 'Other': 0, 'Abroad':0})
df
```

```python
df['corona_result'] = df['corona_result'].map({'positive': 1, 'negative': 0, 'other':0})
df
```

```python
df['age_60_and_above'] = df['age_60_and_above'].map({'No':0, 'Yes':1})
df
```

```python
df = df.dropna()
df
```

```python
df['age_60_and_above'] = df['age_60_and_above'].astype(int)
```

```python
df.to_csv('March_April_00833_processed.csv')
```
